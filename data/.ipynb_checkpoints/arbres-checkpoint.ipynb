# Extract all in rows and save to list

```python
import os
import csv
with open("datatree.csv", "r", encoding="utf-8") as file:
    alldata = csv.reader(file)
    list_alldata=list(alldata)
```

# Nb trees in dataset

```python
list_tree=list_alldata[1:]
nb_tree=len(list_tree)
nb_tree
```

```python
idx_aplantation=list_alldata[0].index('ANNEEDEPLANTATION')
```

# Mean Nb trees by year

```python
minYear=list_alldata[1][idx_aplantation]
maxYear=list_alldata[1][idx_aplantation]
for line in list_alldata[1:]:
    year=line[idx_aplantation]
    if year<minYear and year!="":
        minYear=year
    if year>maxYear and year!="":
        maxYear=year

mean_tree=int(nb_tree)/(int(maxYear)-int(minYear))
        
print(minYear)
print(maxYear)
print(mean_tree)
```

# Nb tree planted in 1986 

```python
list_tree_birthday=[]
for line in list_alldata:
    year=line[idx_aplantation]
    if year=="1986":
        list_tree_birthday.append(line)

print(len(list_tree_birthday))
```

# Nb tree planted since 1986

```python
#Manque if pour nettoyer dataset des valeurs =''
list_tree_since_birthday=[]
for line in list_alldata:
    year=line[idx_aplantation]
    if year<="1986":
        list_tree_since_birthday.append(line)

print(len(list_tree_since_birthday))
```

# Matplotlib Nb trees by year

```python
from collections import Counter

with open('datatree.csv', newline='') as csvfile:
    alltree = csv.DictReader(csvfile,delimiter=',')
    cnt=Counter()
    for tree in alltree:
        if(tree['ANNEEDEPLANTATION'] != ''):
            if tree['ANNEEDEPLANTATION'] in cnt.keys():
                cnt[tree['ANNEEDEPLANTATION']]+=1
            else:
                cnt[tree['ANNEEDEPLANTATION']]=1

years=list(cnt)
print(years)

nbtree=[]
for year in years:
    nbtree.append(cnt[year])
print(nbtree)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x,y=zip(*sorted(cnt.items()))
locs=range(len(x))

xlabel("Années")
ylabel("Nbre d'arbres plantés")
xticks(np.arange(0, 100, step=50))
plt.bar(x,y)
```

```python
fig = plt.figure()
fig.suptitle("Evolution de la plantation d'arbres depuis 1950")
# fig, ax_lst = plt.subplots(2, 3)
fig, ax = plt.subplots()

# xlabel("Années")
# ylabel("Nbre d'arbres plantés")

ax.bar(x, y)
ax.set_xlabel("Années")
ax.set_ylabel("Nbre d'arbres plantés")
ax.set_xticks(np.arange(0, 100, step=50))

plt.show()
```
